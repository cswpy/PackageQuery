```python
import numpy as np

f = open('backup/Z1_critical4.csv', 'r')
dic = {}
hs = list(i for i in range(1, 16, 2))
for line in f:
    ls = line.split(',')
    tokens = ls[0].split("_")
    g = float(tokens[2][1:].replace('d', '.'))
    s = float(tokens[3][1:])
    n = tokens[0]
    if '*' in n:
        g = str(g)+"*"
    if (g, s) not in dic:
        dic[(g, s)] = {"ptime": []}
        for h in hs:
            dic[(g, s)][int(h)] = [[], []]
    if len(ls) == 2:  
        ptime = float(ls[1])
        dic[(g, s)]["ptime"].append(ptime)
    elif len(ls) == 3:
        h = int(float(ls[1]))
        v = float(ls[2])
        c = tokens[-1]
        if c == "time":
            dic[(g, s)][h][0].append(v)
        elif c == "err":
            dic[(g, s)][h][1].append(v)

for (g,s) in dic:
    print("GS:", (g,s))
    print("PTIME", np.mean(dic[(g,s)]['ptime']), len(dic[(g,s)]['ptime']))
    for h in hs:
        print("TIME H="+str(h), np.mean(dic[(g,s)][h][0]), len(dic[(g,s)][h][0]))
        print("ERR H="+str(h), np.mean(dic[(g,s)][h][1]), len(dic[(g,s)][h][1]))
```
