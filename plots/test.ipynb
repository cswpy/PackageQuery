```python
r = Reader("DLV.csv")
curves = [r.get_curve(i) for i in [0, 2]]
curves = [(X[:], np.array(Y)[:]) for (X,Y) in curves]
# print(curves)
n = X[0]

import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 5
#define colors to use
col1 = 'steelblue'
col2 = 'red'

#define subplots
fig,ax = plt.subplots()
ftsize = 18

#add first line to plot
ax.scatter(curves[0][0], curves[0][1], color=col1, marker='.', linewidth=2, s=4)

#add x-axis label
ax.set_xlabel('Bounding variance $\it{}$', fontsize=ftsize)

#add y-axis label
ax.set_ylabel('Number of partition subsets $\it{p}$', color=col1, fontsize=ftsize)

ax.set_yscale("log")
ax.set_xscale("log")

#define second y-axis that shares x-axis with current plot
ax2 = ax.twinx()

#add second line to plot
ax2.scatter(curves[1][0], curves[1][1], color=col2, marker='.', linewidth=2, s=4)

#add second y-axis label
ax2.set_ylabel('RWC score $\it{v_{RWC}}$', color=col2, fontsize=ftsize)
# plt.xticks(fontsize=20)
ax2.yaxis.set_tick_params(labelsize=ftsize)
ax.yaxis.set_tick_params(labelsize=ftsize)
ax.xaxis.set_tick_params(labelsize=ftsize)
```
